```python
# ─── 0) Install dependencies (if not already done) ─────────────────────────────
# In your notebook or terminal:
# !pip install -q pandas sentence-transformers elasticsearch

import pandas as pd
from sentence_transformers import SentenceTransformer
from elasticsearch import Elasticsearch, helpers

# ─── 1) LOAD YOUR NORMAL_USER DATAFRAME ────────────────────────────────────────
# Replace this with the path to your CSV/Excel/whatever. For example:
# df_normal = pd.read_csv("normal_users.csv")

# For demonstration, here’s a toy DataFrame with the required columns:

df_normal = df1

# 1a) Ensure all required columns exist
required_cols = [
    "uid", "first_name", "last_name", "degree", "department", "specialization",
    "skills", "designation", "industry", "institute", "company", "user_type"
]
missing = [c for c in required_cols if c not in df_normal.columns]
if missing:
    raise ValueError(f"Missing columns in normal_user DataFrame: {missing}")

# 1b) Replace any NaN in string fields with empty string
for col in required_cols:
    df_normal[col] = df_normal[col].fillna("").astype(str)

# ─── 2) CREATE TEXT BLOBS FOR EMBEDDING ────────────────────────────────────────
df_normal["academic_text"]     = (
    df_normal["degree"] + " "
    + df_normal["department"] + " "
    + df_normal["specialization"]
)

df_normal["professional_text"] = (
    df_normal["skills"] + " "
    + df_normal["designation"] + " "
    + df_normal["industry"]
)

# For normal_user, there is no mentorship topic or mentee_problem, so leave empty:
df_normal["topic_text"]        = ""

# ─── 3) LOAD THE JOBBERT MODEL ─────────────────────────────────────────────────
model = SentenceTransformer("TechWolf/JobBERT-v2")

# 3a) Verify that calling encode on an empty string returns a 768‐dim vector
_zero_vec = model.encode("", convert_to_numpy=True)
assert len(_zero_vec) == 768

# ─── 4) GENERATE & STORE VECTORS IN THE DATAFRAME ──────────────────────────────
academic_vecs     = []
professional_vecs = []
topic_vecs        = []

for _, row in df_normal.iterrows():
    acad_emb = model.encode(row["academic_text"],     convert_to_numpy=True)
    prof_emb = model.encode(row["professional_text"], convert_to_numpy=True)
    top_emb  = model.encode(row["topic_text"],        convert_to_numpy=True)  # all zeros or encoding of ""

    academic_vecs.append(acad_emb)
    professional_vecs.append(prof_emb)
    topic_vecs.append(top_emb)

df_normal["academic_vec"]     = academic_vecs
df_normal["professional_vec"] = professional_vecs
df_normal["topic_vec"]        = topic_vecs
```

```python

# Convert to JSON-serializable lists:
df_normal["academic_vec_list"]     = df_normal["academic_vec"].apply(lambda x: x.tolist())
df_normal["professional_vec_list"] = df_normal["professional_vec"].apply(lambda x: x.tolist())
df_normal["topic_vec_list"]        = df_normal["topic_vec"].apply(lambda x: x.tolist())

# ─── 5) CONNECT TO ELASTICSEARCH ───────────────────────────────────────────────
# ES_HOST   = "https://a7b16fc079fb47698fe7b89b50ff406b.us-east-1.aws.found.io:443"
# API_KEY   = "OTNhN0lKY0IxaWdKNjRyOXg4UUY6WVZKTW5aX3VuQVRtUEhrVDVNWWxMWWxMQQ=="
# INDEX     = "version-1"

ES_HOST = "https://a7b16fc079fb47698fe7b89b50ff406b.us-east-1.aws.found.io:443"
API_KEY = "OTNhN0lKY0IxaWdKNjRyOXg4UUY6WVZKTW5aX3VuQVRtUEhrVDVNWWxMQQ=="
INDEX = "version-1"

es = Elasticsearch(ES_HOST, api_key=API_KEY)

# ─── 6) VERIFY INDEX MAPPING ───────────────────────────────────────────────────
# (Ensure mapping already has these dense_vector fields: academic_vec, professional_vec, mentorship_topic_vec)
mapping = es.indices.get_mapping(index=INDEX)[INDEX]["mappings"]["properties"]
expected_vec_fields = ["academic_vec", "professional_vec", "mentorship_topic_vec"]
for vf in expected_vec_fields:
    if vf not in mapping or mapping[vf]["type"] != "dense_vector":
        raise RuntimeError(f"Index mapping missing or incorrect for field: {vf}")

# ─── 7) PREPARE BULK ACTIONS & INDEX ───────────────────────────────────────────
def generate_normaluser_actions(df, index_name):
    """
    Yields a dict per row in the format required by helpers.bulk(). Each document:
      - _index: index_name
      - _id:    uid
      - _source: all fields matching the ES mapping
    """
    for _, row in df.iterrows():
        yield {
            "_index": index_name,
            "_id":    row["uid"],
            "_source": {
                "uid":                  row["uid"],
                "first_name":           row["first_name"],
                "last_name":            row["last_name"],
                "degree":               row["degree"],
                "department":           row["department"],
                "specialization":       row["specialization"],
                "skills":               row["skills"],
                "designation":          row["designation"],
                "company":              row["company"],
                "industry":             row["industry"],
                "institute":            row["institute"],
                "user_type":            row["user_type"],   # "normal_user"
                "topic":                "",                  # leave blank or
                                                             # you could store row["topic_text"]
                # Dense vectors (as lists):
                "academic_vec":         row["academic_vec_list"],
                "professional_vec":     row["professional_vec_list"],
                "mentorship_topic_vec": row["topic_vec_list"]
            }
        }

# Execute the bulk indexing
helpers.bulk(es, generate_normaluser_actions(df_normal, INDEX))
es.indices.refresh(index=INDEX)

print(f"✅ Indexed {len(df_normal)} normal_user documents into '{INDEX}'")
```
