```python
import pandas as pd
from sentence_transformers import SentenceTransformer

# 2) Load the CSV into a DataFrame
#    Replace "path/to/your.csv" with the path or file you have.
df = pd.read_csv("demo_standarised.csv")

# 3) Rename columns for clarity
df = df.rename(columns={
    "First_Name":            "first_name",
    "Last_Name":             "last_name",
    "Course/Degree":         "degree",
    "dep":                   "department",
    "Specialization":        "specialization",
    "Skills":                "skills",
    "Current Designation":   "designation",
    "Current Company":       "company",
    "Work Industry":         "industry",
    "Institute":             "institute",
    "Unique Profile ID":     "uid",
    "mentee_problem":        "mentee_problem",
    "mentor_topic":          "mentor_topic",
    "user_type":             "user_type"
})

# 4) Create three new textâ€columns for embedding:
#    â€¢ academic_text     = "degree department specialization"
#    â€¢ professional_text = "skills designation industry"
#    â€¢ topic_text        = either mentee_problem or mentor_topic depending on user_type

# Ensure no NaNs break string concatenation:
df["degree"]          = df["degree"].fillna("")
df["department"]      = df["department"].fillna("")
df["specialization"]  = df["specialization"].fillna("")
df["skills"]          = df["skills"].fillna("")
df["designation"]     = df["designation"].fillna("")
df["industry"]        = df["industry"].fillna("")
df["mentee_problem"]  = df["mentee_problem"].fillna("")
df["mentor_topic"]    = df["mentor_topic"].fillna("")
df["user_type"]       = df["user_type"].fillna("")

df["academic_text"] = (
    df["degree"] + " "
  + df["department"] + " "
  + df["specialization"]
)

df["professional_text"] = (
    df["skills"] + " "
  + df["designation"] + " "
  + df["industry"]
)

# Choose topic_text based on user_type
def pick_topic(row):
    if str(row["user_type"]).lower() == "mentee":
        return row["mentee_problem"]
    else:
        return row["mentor_topic"]

df["topic_text"] = df.apply(pick_topic, axis=1).fillna("")

# 5) Load the JobBERT model once
model = SentenceTransformer("TechWolf/JobBERT-v2")

# 6) Generate the three embeddings row by row
#    Weâ€™ll store them in new DataFrame columns (`academic_vec`, `professional_vec`, `topic_vec`).
academic_vecs = []
professional_vecs = []
topic_vecs = []

for idx, row in df.iterrows():
    acad_str = row["academic_text"]
    prof_str = row["professional_text"]
    top_str  = row["topic_text"]
    
    acad_emb = model.encode(acad_str,    convert_to_numpy=True)
    prof_emb = model.encode(prof_str,    convert_to_numpy=True)
    top_emb  = model.encode(top_str,     convert_to_numpy=True)
    
    academic_vecs.append(acad_emb)
    professional_vecs.append(prof_emb)
    topic_vecs.append(top_emb)

df["academic_vec"]     = academic_vecs
df["professional_vec"] = professional_vecs
df["topic_vec"]        = topic_vecs

# 7) If you need JSONâ€serializable lists (for Elasticsearchâ€™s dense_vector),
#    convert each numpy array to a Python list:
df["academic_vec_list"]     = df["academic_vec"].apply(lambda x: x.tolist())
df["professional_vec_list"] = df["professional_vec"].apply(lambda x: x.tolist())
df["topic_vec_list"]        = df["topic_vec"].apply(lambda x: x.tolist())
```

```python
from elasticsearch import Elasticsearch, helpers
import pandas as pd

# â”€â”€â”€ 1) YOUR DATAFRAME SHOULD EXIST ALREADY â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
#    It must contain these columns:
#      ['uid', 'first_name', 'last_name', 'degree', 'department', 'specialization',
#       'skills', 'designation', 'company', 'industry', 'institute', 'user_type',
#       'topic_text', 'academic_vec_list', 'professional_vec_list', 'topic_vec_list']
#
#    If it's not loaded, uncomment & modify the next line to point to your CSV:
# df = pd.read_csv("path/to/your_processed.csv")

#  Verify that the embedding columns actually have lists of length 768
string_cols = [
    "first_name", "last_name", "degree", "department", "specialization",
    "skills", "designation", "company", "industry", "institute",
    "user_type", "topic_text"
]

for col in string_cols:
    if col in df.columns:
        df[col] = df[col].fillna("").astype(str)

def is_valid_vector(vec):
    return isinstance(vec, list) and len(vec) == 768 and all(isinstance(x, (float, int)) for x in vec)

bad_rows = []
for i, row in df.iterrows():
    if not is_valid_vector(row.get("academic_vec_list", [])):
        bad_rows.append((row["uid"], "academic_vec_list"))
    if not is_valid_vector(row.get("professional_vec_list", [])):
        bad_rows.append((row["uid"], "professional_vec_list"))
    if not is_valid_vector(row.get("topic_vec_list", [])):
        bad_rows.append((row["uid"], "topic_vec_list"))

if bad_rows:
    print("âš ï¸  Found invalid embedding vectors in these rows (uid, field):")
    for uid, field_name in bad_rows:
        print(f"   â€¢ uid={uid}, invalid field={field_name}")
    raise SystemExit("Fix the above rows vectors before indexing.")

# â”€â”€â”€ 2) CONNECT TO ELASTICSEARCH â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
ES_HOST = "https://a7b16fc079fb47698fe7b89b50ff406b.us-east-1.aws.found.io:443"
API_KEY = "OTNhN0lKY0IxaWdKNjRyOXg4UUY6WVZKTW5aX3VuQVRtUEhrVDVNWWxMQQ=="
INDEX_NAME = "version-1"

es = Elasticsearch(
    ES_HOST,
    api_key=API_KEY
)

# â”€â”€â”€ 3) DEFINE & CREATE THE INDEX MAPPING â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
mapping = {
    "mappings": {
        "properties": {
            "uid":                  { "type": "keyword" },
            "first_name":           { "type": "text"    },
            "last_name":            { "type": "text"    },
            "degree":               { "type": "text"    },
            "department":           { "type": "text"    },
            "specialization":       { "type": "text"    },
            "skills":               { "type": "text"    },
            "designation":          { "type": "text"    },
            "company":              { "type": "text"    },
            "industry":             { "type": "text"    },
            "institute":            { "type": "text"    },
            "user_type":            { "type": "keyword" },
            "topic":                { "type": "text"    },
            "academic_vec":         { "type": "dense_vector", "dims": 768 },
            "professional_vec":     { "type": "dense_vector", "dims": 768 },
            "mentorship_topic_vec": { "type": "dense_vector", "dims": 768 }
        }
    }
}

# If you need to recreate from scratch, uncomment these lines:
# if es.indices.exists(INDEX_NAME):
#     es.indices.delete(index=INDEX_NAME)
es.indices.create(index=INDEX_NAME, body=mapping, ignore=400)

# â”€â”€â”€ 4) PREPARE DOCUMENTS FOR BULK INDEXING â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def generate_actions(dataframe, index_name):
    for _, row in dataframe.iterrows():
        uid = row["uid"]
        # Ensure UID is not null or empty
        if pd.isna(uid) or uid == "":
            print(f"âš ï¸  Skipping row with missing UID at index={_}")
            continue

        yield {
            "_index": index_name,
            "_id":    str(uid),
            "_source": {
                "uid":                  str(uid),
                "first_name":           row["first_name"],
                "last_name":            row["last_name"],
                "degree":               row["degree"],
                "department":           row["department"],
                "specialization":       row["specialization"],
                "skills":               row["skills"],
                "designation":          row["designation"],
                "company":              row["company"],
                "industry":             row["industry"],
                "institute":            row["institute"],
                "user_type":            row["user_type"],
                "topic":                row["topic_text"],
                "academic_vec":         row["academic_vec_list"],
                "professional_vec":     row["professional_vec_list"],
                "mentorship_topic_vec": row["topic_vec_list"]
            }
        }

# â”€â”€â”€ 5) EXECUTE BULK INDEXING WITH ERROR CATCHING â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
try:
    success, errors = helpers.bulk(es, generate_actions(df, INDEX_NAME), raise_on_error=False, stats_only=False)
    es.indices.refresh(index=INDEX_NAME)
    print(f"âœ… Successfully indexed {success} documents into '{INDEX_NAME}'.")
    if errors:
        print("âš ï¸  The following errors occurred during bulk indexing:")
        for err in errors:
            # err is a dict: {<op_type>: { "error": { ... }, "_id": <id> } }
            # e.g. { "index": { "_id":"123", "status":400, "error":{...} } }
            op_type, details = next(iter(err.items()))
            doc_id = details.get("_id")
            err_msg = details.get("error", {})
            print(f"   â€¢ doc_id={doc_id}, operation={op_type}, error={err_msg}")
except Exception as e:
    print("âŒ Bulk indexing failed with exception:")
    print(str(e))
    raise
```
